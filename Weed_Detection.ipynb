```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image_path = 'plant.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Convert to grayscale

# Apply Gaussian blur to reduce noise and enhance edge detection
blurred = cv2.GaussianBlur(image, (5, 5), 0)

# Apply Canny edge detection
edges = cv2.Canny(blurred, threshold1=30, threshold2=70)  # Adjust thresholds as needed

# Display the original image and the detected edges
plt.subplot(121), plt.imshow(image, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(edges, cmap='gray')
plt.title('Canny Edges'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('weed.jpeg')
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define the lower and upper bounds of green color for weed detection
lower_green = np.array([35, 50, 50])  # HSV values
upper_green = np.array([85, 255, 255])

# Create a mask for green color
green_mask = cv2.inRange(hsv_image, lower_green, upper_green)

# Find contours of detected green regions
contours, _ = cv2.findContours(green_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image
output_image = image.copy()
cv2.drawContours(output_image, contours, -1, (0, 255, 0), 2)  # Green color for contours

# Display the original image with detected weeds
plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
plt.title('Detected Weeds')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('plant.jpg', cv2.IMREAD_GRAYSCALE)

# Define the Sobel filter for edge detection
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])
sobel_y = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])

# Apply convolution to the image using the Sobel filters
edges_x = cv2.filter2D(image, cv2.CV_64F, sobel_x)
edges_y = cv2.filter2D(image, cv2.CV_64F, sobel_y)

# Calculate the gradient magnitude
gradient_magnitude = np.sqrt(edges_x**2 + edges_y**2)

# Display the original image and gradient magnitude
plt.subplot(131), plt.imshow(image, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(np.abs(edges_x), cmap='gray')
plt.title('Edges in X direction'), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Gradient Magnitude'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('weed.jpeg', cv2.IMREAD_GRAYSCALE)

# Define the translation matrix
dx, dy = 20, 30  # Shift values (pixels)
translation_matrix = np.float32([[1, 0, dx],
                                 [0, 1, dy]])

# Apply translation to the image
translated_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))

# Display the original and translated images
plt.subplot(121), plt.imshow(image, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(translated_image, cmap='gray')
plt.title('Translated Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read the image
input_image = cv2.imread("plant.jpg")

# Define the rotation angle (in degrees)
angle = 30

# Get the image dimensions
height, width = input_image.shape[:2]

# Calculate the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1)

# Apply the rotation
rotated_image = cv2.warpAffine(input_image, rotation_matrix, (width, height))

# Display the original and rotated images
cv2_imshow(input_image)
cv2_imshow(rotated_image)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read the image
input_image = cv2.imread("plant.jpg")

# Define shear parameters
shear_factor_x = 0.3
shear_factor_y = 0.1

# Get the image dimensions
height, width = input_image.shape[:2]

# Define the shear matrix
shear_matrix = np.array([[1, shear_factor_x, 0],
                         [shear_factor_y, 1, 0]])

# Apply the shear transformation
sheared_image = cv2.warpAffine(input_image, shear_matrix, (width, height))

# Display the original and sheared images
cv2_imshow(input_image)
cv2_imshow(sheared_image)
```
